```python
import cryptocode
import random
import os
```

```python
_chars = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%&*()-_=+'
```

```python
key = ''.join([_chars[random.randint(0, len(_chars) - 1)] for _ in range(len(_chars))])
```

```python
enc_user = cryptocode.encrypt('U0000753',key)
enc_pass = cryptocode.encrypt('@H1MuR4_2025_s4',key)
```

```python

def criar_arquivo_criptografado(user, password, path = os.getcwd()):
    enc_user = cryptocode.encrypt(user, key)
    enc_pass = cryptocode.encrypt(password, key)

    # Escreve uma chave aleatória para criptografia
    with open(path + '\\' + 'key', 'w') as _key:
        _key.write(key)

    # Escreve uma chave aleatória para criptografar o >> USUÁRIO <<
    with open(path + '\\' + 'user', 'w') as _userfile:
        _userfile.write(enc_user)
    
    # Escreve uma chave aleatório para criptografar a >> SENHA <<
    with open(path + '\\' + 'senha', 'w') as _passfile:
        _passfile.write(enc_pass)
```

```python
criar_arquivo_criptografado("user", crypt = enc_user)
```
